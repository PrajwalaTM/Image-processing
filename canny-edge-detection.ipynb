```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
image = cv2.imread('Edge2.jpeg',0)
```

```python
plt.imshow(image,"gray"),plt.title("Original")
plt.show()
```

```python
canny_image1 = cv2.Canny(image,100,200,1)
canny_image2 = cv2.Canny(image,100,200)
```

```python
plt.imshow(canny_image1,"gray"),plt.title("Edge detection")
plt.show()
```

```python
plt.imshow(canny_image2,"gray"),plt.title("Edge detection")
plt.show()
```
