```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('5.png',0)
```

```python
plt.imshow(img,"gray")
plt.show()
```

```python
kernel = np.ones((10,10),np.uint8)
eroded = cv2.erode(img,kernel,15)
dilated = cv2.dilate(img,kernel,15)
```

```python
plt.subplot(131),plt.imshow(img,"gray"),plt.title("original")
plt.subplot(132),plt.imshow(eroded,"gray"),plt.title("eroded")
plt.subplot(133),plt.imshow(dilated,"gray"),plt.title("dilated")
plt.show()
```

```python
img1 = cv2.imread('opening.png')
img2 = cv2.imread('closing.png')
```

```python
kernel= np.ones((5,5),np.uint8)
opened_img = cv2.morphologyEx(img1,cv2.MORPH_OPEN,kernel)
closed_img = cv2.morphologyEx(img2,cv2.MORPH_CLOSE,kernel)
```

```python
plt.subplot(121),plt.imshow(img1),plt.title("original")
plt.subplot(122),plt.imshow(opened_img),plt.title("Opened image")
plt.show()
```

```python
plt.subplot(121),plt.imshow(img2),plt.title("original")
plt.subplot(122),plt.imshow(closed_img),plt.title("Closed image")
plt.show()
```

```python
# boundary of the image - difference between opening and closing
kernel = np.ones((2,2),np.uint8)
gradient_img = cv2.morphologyEx(img,cv2.MORPH_GRADIENT,kernel)
```

```python
plt.imshow(gradient_img,"gray"),plt.title("Boundary")
plt.show()
```

```python
cameraman = cv2.imread('cameraman.png',0)
```

```python
laplacian_filter = cv2.Laplacian(cameraman,cv2.CV_64F)
```

```python
sobel_x  = cv2.Sobel(cameraman,cv2.CV_64F,1,0,ksize=1)
sobel_y = cv2.Sobel(cameraman,cv2.CV_64F,0,1,ksize=1)
```

```python
plt.subplot(141),plt.imshow(cameraman,"gray"),plt.title("original")
plt.subplot(142),plt.imshow(laplacian_filter,"gray"),plt.title("laplacian")
plt.subplot(143),plt.imshow(sobel_x,"gray"),plt.title("sobel_x")
plt.subplot(144),plt.imshow(sobel_y,"gray"),plt.title("sobel_y")
plt.show()
```
