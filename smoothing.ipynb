```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('cat.jpg',0)
```

```python
#averaging filter
kernel = np.ones((5,5),np.float32)/25
smooth_img = cv2.filter2D(img,-1,kernel) 
```

```python
plt.subplot(121),plt.imshow(img,"gray")
plt.subplot(122),plt.imshow(smooth_img,"gray")
plt.show()
```

```python
#box filter
blur_img = cv2.blur(img,(5,5))
```

```python
plt.subplot(121),plt.imshow(img,"gray")
plt.subplot(122),plt.imshow(blur_img,"gray")
plt.show()
```

```python
#gaussian blur
gaussian_blur = cv2.GaussianBlur(img,(5,5),0)
```

```python
plt.subplot(121),plt.imshow(img,"gray")
plt.subplot(122),plt.imshow(gaussian_blur,"gray")
plt.show()
```

```python
#median blur
median_blur = cv2.medianBlur(img,5)
```

```python
plt.subplot(121),plt.imshow(img,"gray")
plt.subplot(122),plt.imshow(median_blur,"gray")
plt.show()
```

```python
#NLM for denoising
img = cv2.imread('Guassian.jpeg')
plt.imshow(img,"gray")
plt.show()
```

```python
denoised_img = cv2.fastNlMeansDenoising(img,None,10,7,21)
plt.imshow(denoised_img,"gray")
plt.show()
```
